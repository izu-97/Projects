```python
url = "https://ssd-api.jpl.nasa.gov/fireball.api?date-min=2000-01-01&req-alt=true"
```

```python
r = requests.get(url)
```

```python
df.columns = index
```

```python
df["date"] = df["date"].str.slice(stop=10)
```
